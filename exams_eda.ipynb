# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/exams.csv')
df.head()
```

```python
df.info()
```

Data is very clean.

```python
df.describe()
```

Seems like people are generally worse at math compared to reading and writting. The minimum score for math is **14 points** and **10 points** lower than reading and writing respectively.

**Idea:** And |the average is lower too. An hypothesis test (t-test) can be done to see if math scores really are different from reading and writing scores.

```python
# split continuous and categorical data
df_cont = df[['reading score', 'writing score', 'math score']].copy()
df_cat = df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'math score']].copy()
```

## Continuous Data

**Question**: Are there any correlation between math scores and reading/writting scores?

```python
fig, ax = plt.subplots(figsize=(4, 4))
sns.scatterplot(
    data=df_cont, 
    x='reading score', 
    y='math score', 
    hue='writing score',
    palette=sns.color_palette("flare_r", as_cmap=True),
    ax=ax
)

ax.set_title('Student Reading/Writing vs Math Score')
ax.set_xlabel('Reading Score')
ax.set_ylabel('Math Score')

legend = ax.legend()

norm = plt.Normalize(0, 100)
sm = plt.cm.ScalarMappable(cmap="flare_r", norm=norm)
sm.set_array([])

# Remove the legend and add a colorbar
ax.get_legend().remove()
colorbar = ax.figure.colorbar(sm)
colorbar.set_label('Writing Score', rotation=270);
```

It seems the better a person is at reading and writing the better they would be at math as well.

Let's quatify this correlation.

```python
sns.heatmap(data=df_cont.corr(), annot=True);
```

**Insight**: writing score and reading score have a high correlation, we probably don't need both of them as features.

math score and reading/writting score correlation p-value:

```python
from scipy.stats import pearsonr

corr, p = pearsonr(df_cont['reading score'], df_cont['math score'])
print(f'Reading vs Math score: corr={corr:.2f} ({p:.3f})')

corr, p = pearsonr(df_cont['writing score'], df_cont['math score'])
print(f'Writing vs Math score: corr={corr:.2f} ({p:.3f})')
```

The correlations are significant as the p-values are less than 0.05

**Conclusion**: Math and Reading/Writing are highly correlated with a personr value of 0.82 and 0.81 respectively and p-value less than 0.05. This makes sense, as Math would require similar language comprehension skills as reading and writing would.

**Future Experiments**: It would be interesting to see if a person's bilingualism affects their math scores or not.

## Categorical Data

```python
df_cat.head()
```

**Question**: Are male students better than female students at math?

```python
df_cat.gender.value_counts()
```

There are similar quantity of data avalilable for male and female math scores, 517 : 483.

```python
df_grouped = df_cat.groupby('gender').mean()
ax = df_grouped.plot(kind='bar')

ax.get_legend().remove()
ax.set_ylabel('Math Score')
ax.set_xlabel('Gender')

ax.set_ylim(0, 80)

ax.annotate(f"{df_grouped.loc['female', 'math score']:.2f}", xy=(0, 65), xytext=(-10, 0), textcoords='offset points')
ax.annotate(f"{df_grouped.loc['male', 'math score']:.2f}", xy=(1, 70), xytext=(-10, 5), textcoords='offset points');
```

```python
sns.violinplot(data=df_cat, x='gender', y='math score');
```

the average math scores for female are 6 points lower than that of males. Is this statistically significant? A t-test needs to be performed. From the violin plot it seems that scores at the lower end were from female students.

```python
# isolate female and male math scores
female_scores = df_cat.loc[df_cat['gender'] == 'female', 'math score'].values
male_scores = df_cat.loc[df_cat['gender'] == 'male', 'math score'].values
```

```python
female_score_std = female_scores.std()
male_score_std = male_scores.std()

print(f'Standard dev of female vs male math score: ({female_score_std:.2f}, {male_score_std:.2f})')
```

```python
from scipy.stats import ttest_ind

t_stat, p = ttest_ind(male_scores, female_scores, equal_var=False, alternative='greater')

print(f't_stat={t_stat:.3f} ({p:.2f})')
```

Since the p-value < 0.05, the difference between mean male and female students math scores are statistically significant.

**Conclusion**: The data does seem to provide some evidence towards, males performing slighly better than female students on average in Math.

## Data Clustering

**Question**: Can we create any clustering based on the categorical and continous features?

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(8, 8))

gs = fig.add_gridspec(3, 2)

axes = [
    fig.add_subplot(gs[0, :]), 
    fig.add_subplot(gs[1, 0]), 
    fig.add_subplot(gs[1, 1]), 
    fig.add_subplot(gs[2, 0]), 
    fig.add_subplot(gs[2, 1])
]

X = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for x, ax in zip(X, axes):
    df_sorted = df.sort_values(by=x)
    sns.scatterplot(data = df_sorted, x='writing score', y='math score', ax=ax, hue=x)

plt.tight_layout();
```

**Conclusion**: There seems to be no clear cut clustering to be observed. Again the scatter plot for gender seems to show male students performing better than female students.
